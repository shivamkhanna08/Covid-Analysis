## Data Visualization 

EXPLORATORY DATA ANALYSIS

```python
#Data distribution of all the numeric variables
%matplotlib inline
import matplotlib.pyplot as plt
df.hist(bins=20, figsize=(25,20))
plt.show()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Active Ratio (%)", data=df)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Discharge Ratio (%)", data=df)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Death Ratio (%)", data=df)
plt.show()
```

```python
sns.distplot(df['Active'])
```

```python
sns.distplot(df['Discharged'])
```

```python
df.corr
```

```python
#Correlation matrix of the dataset
f, ax = plt.subplots(figsize= [15,10])
sns.heatmap(df.corr(), annot=True, fmt=".2f", ax=ax)
ax.set_title("Correlation Matrix", fontsize=20) 
plt.show()
```

```python
import matplotlib.pyplot as plt
fig_dims = (22, 11)
fig, ax = plt.subplots(figsize=fig_dims)

#ax = axis
sns.countplot(x='Deaths',hue='Discharged',data=df,palette="colorblind", ax=ax, edgecolor=sns.color_palette("dark", n_colors=1));
```

```python
# Top 5 active cases of states in the country
top_5 = df.sort_values(by=['Active'], ascending=False).head()

plt.figure(figsize=(15,5))
Active = sns.barplot(top_5['Active'], top_5['State/UTs'], color = 'green', label='Confirmed')
```

```python
#Top 5 total cases of states in the country
top_5 = df.sort_values(by=['Total Cases'], ascending=False).head()

plt.figure(figsize=(15,5))
Total = sns.barplot(top_5['Total Cases'], top_5['State/UTs'], color = 'red', label='Total Cases')
```

```python
# Top 5 discharged patients of states in the country
top_5 = df.sort_values(by=['Discharged'], ascending=False).head()

plt.figure(figsize=(15,5))
Active = sns.barplot(top_5['Discharged'], top_5['State/UTs'], color = 'cyan', label='Confirmed')
```

```python
#State/UTs vs Total Cases
plt.figure(figsize=(25,10))
Plot = sns.barplot(x="State/UTs", y="Total Cases", data=df, palette="Set1")
```
